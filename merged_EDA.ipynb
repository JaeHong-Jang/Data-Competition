```python
# is_numeric_dtype 함수를 사용하기 위해 import
from pandas.api.types import is_numeric_dtype

for column in df1.columns:
    plt.figure(figsize=(10, 6))

    # ===== 데이터 타입에 따라 그래프 분기 =====
    # is_numeric_dtype() 함수로 해당 컬럼이 수치형인지 확인
    if is_numeric_dtype(df1[column]):
        # --- 밀도 플롯 (Density Plot) ---
        sns.kdeplot(df1[column], fill=True, lw=3)
        plt.title(f'"{column}" 변수의 분포 (밀도 플롯)', fontsize=16)
        plt.xlabel(column, fontsize=12)
        plt.ylabel('밀도 (Density)', fontsize=12)
        
    else:
        # --- 막대그래프 (Bar Plot) ---
        # 수치형이 아니면 모두 범주형으로 간주하고 countplot을 그립니다.
        sns.countplot(x=df1[column], palette='viridis', hue=df1[column], legend=False)
        plt.title(f'"{column}" 변수의 빈도수 (막대그래프)', fontsize=16)
        plt.xlabel(column, fontsize=12)
        plt.ylabel('데이터 개수 (Count)', fontsize=12)
        # x축 라벨이 길 경우 겹치지 않도록 회전
        plt.xticks(rotation=45) 

    plt.grid(True, linestyle='--', alpha=0.6)
    # 그래프 레이아웃이 잘리지 않도록 조정
    plt.tight_layout() 
    plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 1. CSV 파일에서 데이터 불러오기
df1 = pd.read_csv('df_merged.csv')

# 2. (핵심) 오류 해결: 숫자형 데이터만 선택하여 새로운 데이터프레임 생성
numeric_df = df1.select_dtypes(include=np.number)

# 3. 숫자형 데이터로 상관관계 행렬 계산
corr_matrix = numeric_df.corr()

# 4. 히트맵 시각화
# 히트맵을 그릴 figure의 사이즈 설정
plt.figure(figsize=(20, 16))

# 상관계수의 절댓값이 0.7 미만인 값들은 그래프에 표시하지 않도록 '마스크(mask)' 생성
mask = np.abs(corr_matrix) < 0.7

# seaborn의 heatmap 함수를 사용하여 시각화
sns.heatmap(corr_matrix, 
            annot=False,      # 셀 안에 상관계수 값을 표시하지 않음
            cmap='viridis',   # 색상 팔레트 지정
            mask=mask,        # 위에서 생성한 마스크 적용
            linewidths=.5)    # 각 셀을 구분하는 라인 추가

# 그래프 제목 및 축 레이블 설정
plt.title('상관계수 0.7 이상인 변수 쌍 히트맵', fontsize=20)
plt.xticks(rotation=45, ha='right') # x축 라벨이 겹치지 않도록 회전
plt.yticks(rotation=0)
plt.tight_layout() # 그래프 요소들이 겹치지 않도록 자동 조정

# 그래프를 이미지 파일로 저장
plt.savefig('correlation_heatmap.png')

# (선택 사항) 화면에 그래프 표시
plt.show()
```

```python
# ====================================
# 2. 이상치 탐지 함수 정의
# ====================================

def detect_outliers_iqr(df, column):
    """IQR 방법으로 이상치 탐지"""
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# ====================================
# 3. df1의 전체 변수에 대한 이상치 탐지 실행
# ====================================

print("="*60)
print("🎯 df1(SSP1)의 전체 변수에 대한 이상치 탐지 (IQR 방법)")
print("="*60)

found_outlier = False

# df1의 모든 열(변수)에 대해 반복 (month는 제외)
for col in df1.columns:
    if col in ['month', 'scenario']:
        continue # month 열은 건너뛰기

    # 이상치 탐지 함수 호출
    outliers, lower, upper = detect_outliers_iqr(df1, col)
    outlier_count = len(outliers)
    
    # 이상치가 발견된 경우에만 결과 출력
    if outlier_count > 0:
        outlier_percentage = (outlier_count / len(df1)) * 100
        print(f"  • {col}: {outlier_count}개 ({outlier_percentage:.1f}%)")
        print(f"    정상범위: [{lower:.2f}, {upper:.2f}]")
        found_outlier = True

if not found_outlier:
    print("  -> 발견된 이상치가 없습니다.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# 한글 폰트 설정
try:
    plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows
except:
    try:
        plt.rcParams['font.family'] = 'AppleGothic'  # Mac
    except:
        plt.rcParams['font.family'] = 'DejaVu Sans'  # Linux
        
plt.rcParams['axes.unicode_minus'] = False

# 2. (핵심) 오류 해결: 숫자형 데이터만 선택하여 새로운 데이터프레임 생성
numeric_df = df1.select_dtypes(include=np.number)

def plot_monthly_avg_histogram(data, col):
    """월별 평균변화 히스토그램 형태로 그리기"""
    monthly_avg = data.groupby('month')[col].mean()
    
    plt.figure(figsize=(10, 6))
    plt.bar(monthly_avg.index, monthly_avg.values, color='skyblue', alpha=0.8)
    plt.title(f'{col} 월별 평균 변화 (히스토그램)', fontsize=14, fontweight='bold')
    plt.xlabel('월', fontsize=12)
    plt.ylabel(f'{col} 평균', fontsize=12)
    plt.grid(axis='y', alpha=0.3)
    plt.xticks(monthly_avg.index)
    plt.tight_layout()
    plt.show()

# 모든 컬럼 월별 평균변화 히스토그램 그리기
print("📊 모든 컬럼 월별 평균변화 히스토그램 생성!")
print(f"총 {len(df1.columns)}개 컬럼 분석")

for i, column in enumerate(numeric_df.columns):
    print(f"\n[{i+1}/{len(df1.columns)}] {column} 월별 평균변화 생성 중...")
    plot_monthly_avg_histogram(numeric_df, column)

print("✅ 모든 컬럼 월별 평균변화 히스토그램 완료!")
```
