```python
# ====================================
# 2. Í≤∞Ï∏°Ïπò Î∂ÑÏÑù
# ====================================

print("="*60)
print("üîç Í≤∞Ï∏°Ïπò Î∂ÑÏÑù")
print("="*60)

def analyze_missing_values(df, name):
    """Í≤∞Ï∏°Ïπò Î∂ÑÏÑù Ìï®Ïàò"""
    total_missing = df.isnull().sum().sum()
    total_cells = df.shape[0] * df.shape[1]
    missing_percentage = (total_missing / total_cells) * 100
    
    print(f"\n[{name} ÏãúÎÇòÎ¶¨Ïò§]")
    print(f"‚Ä¢ Ï†ÑÏ≤¥ Í≤∞Ï∏°Ïπò: {total_missing}Í∞ú ({missing_percentage:.2f}%)")
    
    # Ïó¥Î≥Ñ Í≤∞Ï∏°Ïπò
    missing_by_column = df.isnull().sum()
    missing_columns = missing_by_column[missing_by_column > 0]
    
    if len(missing_columns) > 0:
        print(f"‚Ä¢ Í≤∞Ï∏°ÏπòÍ∞Ä ÏûàÎäî Ïó¥:")
        for col, count in missing_columns.items():
            percentage = (count / len(df)) * 100
            print(f"  - {col}: {count}Í∞ú ({percentage:.2f}%)")
    else:
        print("‚Ä¢ ‚úÖ Í≤∞Ï∏°ÏπòÍ∞Ä ÏóÜÏäµÎãàÎã§!")
    
    return missing_by_column

# Í∞Å Îç∞Ïù¥ÌÑ∞ÏÖãÎ≥Ñ Í≤∞Ï∏°Ïπò Î∂ÑÏÑù
missing_info = {}
for name, df in datasets.items():
    missing_info[name] = analyze_missing_values(df, name)
```

```python
# ====================================
# 4. ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ Î∂ÑÏÑù
# ====================================

print("="*60)
print("üîó ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ Î∂ÑÏÑù")
print("="*60)

# Í∞Å Îç∞Ïù¥ÌÑ∞ÏÖãÎ≥Ñ ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ Îß§Ìä∏Î¶≠Ïä§ Í≥ÑÏÇ∞
correlations = {}
for name, df in datasets.items():
    correlations[name] = df.select_dtypes(include=[np.number]).corr()
```

```python
# ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÌûàÌä∏Îßµ (Í∞Å ÏãúÎÇòÎ¶¨Ïò§Î≥Ñ)
fig, axes = plt.subplots(1, 3, figsize=(20, 6))
fig.suptitle('ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ ÌûàÌä∏Îßµ (SSP ÏãúÎÇòÎ¶¨Ïò§Î≥Ñ)', fontsize=16, fontweight='bold')

for idx, (name, corr) in enumerate(correlations.items()):
    ax = axes[idx]
    
    # Ï£ºÏöî Î≥ÄÏàòÎì§Îßå ÏÑ†ÌÉùÌïòÏó¨ ÌûàÌä∏Îßµ Í∑∏Î¶¨Í∏∞
    important_vars = ['CO2ppm', 'Temp', 'Humid', 'VPD', 'Chl_a', 'TChl', 
                     'PI_abs', 'Fv-Fm', 'Leaf_TPC', 'Root_TPC', 'Leaf_TFC', 'Root_TFC']
    
    # ÏÑ†ÌÉùÌïú Î≥ÄÏàòÎì§Îßå ÌïÑÌÑ∞ÎßÅ
    corr_subset = corr.loc[important_vars, important_vars]
    
    sns.heatmap(corr_subset, annot=True, fmt='.2f', cmap='coolwarm', 
                center=0, ax=ax, cbar_kws={'shrink': 0.8},
                vmin=-1, vmax=1)
    ax.set_title(f'{name} ÏãúÎÇòÎ¶¨Ïò§')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
    ax.set_yticklabels(ax.get_yticklabels(), rotation=0)

plt.tight_layout()
plt.show()
```

```python
# Í∞ïÌïú ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ Ï∞æÍ∏∞
print("\n" + "="*60)
print("üí° Í∞ïÌïú ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ (|r| > 0.7)")
print("="*60)

def find_strong_correlations(corr_matrix, threshold=0.7):
    """Í∞ïÌïú ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÎ•º Ï∞æÎäî Ìï®Ïàò"""
    strong_corr = []
    
    for i in range(len(corr_matrix.columns)):
        for j in range(i+1, len(corr_matrix.columns)):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                strong_corr.append({
                    'var1': corr_matrix.columns[i],
                    'var2': corr_matrix.columns[j],
                    'corr': corr_matrix.iloc[i, j]
                })
    
    return pd.DataFrame(strong_corr).sort_values('corr', key=abs, ascending=False)

# Í∞Å ÏãúÎÇòÎ¶¨Ïò§Î≥Ñ Í∞ïÌïú ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ
for name, corr in correlations.items():
    print(f"\n[{name} ÏãúÎÇòÎ¶¨Ïò§]")
    strong_corr_df = find_strong_correlations(corr, threshold=0.7)
    
    if len(strong_corr_df) > 0:
        for _, row in strong_corr_df.head(10).iterrows():
            sign = "+" if row['corr'] > 0 else "-"
            print(f"  {sign} {row['var1']} ‚Üî {row['var2']}: {row['corr']:.3f}")
    else:
        print("  Í∞ïÌïú ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÎ•º Î≥¥Ïù¥Îäî Î≥ÄÏàòÏåçÏù¥ ÏóÜÏäµÎãàÎã§.")
```

```python
# ====================================
# 5. Ïù¥ÏÉÅÏπò ÌÉêÏßÄ
# ====================================

print("="*60)
print("üéØ Ïù¥ÏÉÅÏπò ÌÉêÏßÄ (IQR Î∞©Î≤ï)")
print("="*60)

def detect_outliers_iqr(df, column):
    """IQR Î∞©Î≤ïÏúºÎ°ú Ïù¥ÏÉÅÏπò ÌÉêÏßÄ"""
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Ï£ºÏöî Î≥ÄÏàòÎì§Ïóê ÎåÄÌïú Ïù¥ÏÉÅÏπò Í≤ÄÏÇ¨
check_columns = ['CO2ppm', 'Temp', 'PI_abs', 'Fv-Fm', 'Leaf_TPC', 'Root_TPC']

outlier_summary = {}
for name, df in datasets.items():
    outlier_summary[name] = {}
    print(f"\n[{name} ÏãúÎÇòÎ¶¨Ïò§ Ïù¥ÏÉÅÏπò]")
    
    for col in check_columns:
        outliers, lower, upper = detect_outliers_iqr(df, col)
        outlier_count = len(outliers)
        outlier_percentage = (outlier_count / len(df)) * 100
        
        outlier_summary[name][col] = {
            'count': outlier_count,
            'percentage': outlier_percentage,
            'lower_bound': lower,
            'upper_bound': upper
        }
        
        if outlier_count > 0:
            print(f"  ‚Ä¢ {col}: {outlier_count}Í∞ú ({outlier_percentage:.1f}%)")
            print(f"    Ï†ïÏÉÅÎ≤îÏúÑ: [{lower:.2f}, {upper:.2f}]")
```

```python
# Ïù¥ÏÉÅÏπò ÏãúÍ∞ÅÌôî
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Ï£ºÏöî Î≥ÄÏàò Ïù¥ÏÉÅÏπò Î∂ÑÌè¨', fontsize=16, fontweight='bold')

for idx, col in enumerate(check_columns):
    ax = axes[idx//3, idx%3]
    
    # Í∞Å ÏãúÎÇòÎ¶¨Ïò§Î≥Ñ Îç∞Ïù¥ÌÑ∞ÏôÄ Ïù¥ÏÉÅÏπò ÌëúÏãú
    for name, df in datasets.items():
        outliers, lower, upper = detect_outliers_iqr(df, col)
        
        # Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞ ÏÇ∞Ï†êÎèÑ
        ax.scatter(df.index, df[col], alpha=0.5, s=10, label=f'{name}')
        
        # Ïù¥ÏÉÅÏπò Í∞ïÏ°∞
        if len(outliers) > 0:
            ax.scatter(outliers.index, outliers[col], color='red', 
                      s=30, marker='x', alpha=0.8)
    
    ax.set_title(f'{col} Ïù¥ÏÉÅÏπò')
    ax.set_xlabel('Index')
    ax.set_ylabel(col)
    ax.legend()
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# ====================================
# 6. SSP ÏãúÎÇòÎ¶¨Ïò§Î≥Ñ ÎπÑÍµê Î∂ÑÏÑù
# ====================================

print("="*60)
print("üîÑ SSP ÏãúÎÇòÎ¶¨Ïò§Î≥Ñ ÎπÑÍµê Î∂ÑÏÑù")
print("="*60)

# ÏãúÎÇòÎ¶¨Ïò§Î≥Ñ ÌèâÍ∑†Í∞í ÎπÑÍµê
comparison_df = pd.DataFrame()

for name, df in datasets.items():
    means = df.select_dtypes(include=[np.number]).mean()
    comparison_df[name] = means

# Ï£ºÏöî Î≥ÄÏàòÎì§Îßå ÏÑ†ÌÉù
important_vars = ['CO2ppm', 'Temp', 'Humid', 'VPD', 'Chl_a', 'TChl', 
                 'PI_abs', 'Fv-Fm', 'Leaf_TPC', 'Root_TPC', 'Leaf_TFC', 'Root_TFC']

comparison_subset = comparison_df.loc[important_vars]

print("\nüìä ÏãúÎÇòÎ¶¨Ïò§Î≥Ñ ÌèâÍ∑†Í∞í ÎπÑÍµê")
print(comparison_subset.round(3))
```

```python
# Î≥ÄÌôîÏú® Í≥ÑÏÇ∞ (SSP1 ÎåÄÎπÑ)
print("\nüìà SSP1 ÎåÄÎπÑ Î≥ÄÌôîÏú® (%)")
change_rate = pd.DataFrame()

baseline = comparison_subset['SSP1']
for scenario in ['SSP3', 'SSP5']:
    change_rate[f'{scenario} vs SSP1'] = ((comparison_subset[scenario] - baseline) / baseline * 100)

print(change_rate.round(2))
```

```python
# ÏãúÎÇòÎ¶¨Ïò§Î≥Ñ ÎπÑÍµê ÏãúÍ∞ÅÌôî
fig, axes = plt.subplots(3, 4, figsize=(20, 15))
fig.suptitle('SSP ÏãúÎÇòÎ¶¨Ïò§Î≥Ñ Ï£ºÏöî Î≥ÄÏàò ÎπÑÍµê', fontsize=16, fontweight='bold')

for idx, var in enumerate(important_vars):
    ax = axes[idx//4, idx%4]
    
    # ÎßâÎåÄ Í∑∏ÎûòÌîÑÎ°ú ÎπÑÍµê
    values = [comparison_subset.loc[var, scenario] for scenario in ['SSP1', 'SSP3', 'SSP5']]
    colors = ['lightblue', 'lightcoral', 'lightgreen']
    bars = ax.bar(['SSP1', 'SSP3', 'SSP5'], values, color=colors, edgecolor='black')
    
    # Í∞í ÌëúÏãú
    for bar, value in zip(bars, values):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height,
                f'{value:.2f}', ha='center', va='bottom', fontsize=9)
    
    ax.set_title(var)
    ax.set_ylabel('ÌèâÍ∑†Í∞í')
    ax.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```
