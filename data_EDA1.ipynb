```python
# ====================================
# 2. 결측치 분석
# ====================================

print("="*60)
print("🔍 결측치 분석")
print("="*60)

def analyze_missing_values(df, name):
    """결측치 분석 함수"""
    total_missing = df.isnull().sum().sum()
    total_cells = df.shape[0] * df.shape[1]
    missing_percentage = (total_missing / total_cells) * 100
    
    print(f"\n[{name} 시나리오]")
    print(f"• 전체 결측치: {total_missing}개 ({missing_percentage:.2f}%)")
    
    # 열별 결측치
    missing_by_column = df.isnull().sum()
    missing_columns = missing_by_column[missing_by_column > 0]
    
    if len(missing_columns) > 0:
        print(f"• 결측치가 있는 열:")
        for col, count in missing_columns.items():
            percentage = (count / len(df)) * 100
            print(f"  - {col}: {count}개 ({percentage:.2f}%)")
    else:
        print("• ✅ 결측치가 없습니다!")
    
    return missing_by_column

# 각 데이터셋별 결측치 분석
missing_info = {}
for name, df in datasets.items():
    missing_info[name] = analyze_missing_values(df, name)
```

```python
# ====================================
# 4. 상관관계 분석
# ====================================

print("="*60)
print("🔗 상관관계 분석")
print("="*60)

# 각 데이터셋별 상관관계 매트릭스 계산
correlations = {}
for name, df in datasets.items():
    correlations[name] = df.select_dtypes(include=[np.number]).corr()
```

```python
# 상관관계 히트맵 (각 시나리오별)
fig, axes = plt.subplots(1, 3, figsize=(20, 6))
fig.suptitle('상관관계 히트맵 (SSP 시나리오별)', fontsize=16, fontweight='bold')

for idx, (name, corr) in enumerate(correlations.items()):
    ax = axes[idx]
    
    # 주요 변수들만 선택하여 히트맵 그리기
    important_vars = ['CO2ppm', 'Temp', 'Humid', 'VPD', 'Chl_a', 'TChl', 
                     'PI_abs', 'Fv-Fm', 'Leaf_TPC', 'Root_TPC', 'Leaf_TFC', 'Root_TFC']
    
    # 선택한 변수들만 필터링
    corr_subset = corr.loc[important_vars, important_vars]
    
    sns.heatmap(corr_subset, annot=True, fmt='.2f', cmap='coolwarm', 
                center=0, ax=ax, cbar_kws={'shrink': 0.8},
                vmin=-1, vmax=1)
    ax.set_title(f'{name} 시나리오')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
    ax.set_yticklabels(ax.get_yticklabels(), rotation=0)

plt.tight_layout()
plt.show()
```

```python
# 강한 상관관계 찾기
print("\n" + "="*60)
print("💡 강한 상관관계 (|r| > 0.7)")
print("="*60)

def find_strong_correlations(corr_matrix, threshold=0.7):
    """강한 상관관계를 찾는 함수"""
    strong_corr = []
    
    for i in range(len(corr_matrix.columns)):
        for j in range(i+1, len(corr_matrix.columns)):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                strong_corr.append({
                    'var1': corr_matrix.columns[i],
                    'var2': corr_matrix.columns[j],
                    'corr': corr_matrix.iloc[i, j]
                })
    
    return pd.DataFrame(strong_corr).sort_values('corr', key=abs, ascending=False)

# 각 시나리오별 강한 상관관계
for name, corr in correlations.items():
    print(f"\n[{name} 시나리오]")
    strong_corr_df = find_strong_correlations(corr, threshold=0.7)
    
    if len(strong_corr_df) > 0:
        for _, row in strong_corr_df.head(10).iterrows():
            sign = "+" if row['corr'] > 0 else "-"
            print(f"  {sign} {row['var1']} ↔ {row['var2']}: {row['corr']:.3f}")
    else:
        print("  강한 상관관계를 보이는 변수쌍이 없습니다.")
```

```python
# ====================================
# 5. 이상치 탐지
# ====================================

print("="*60)
print("🎯 이상치 탐지 (IQR 방법)")
print("="*60)

def detect_outliers_iqr(df, column):
    """IQR 방법으로 이상치 탐지"""
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# 주요 변수들에 대한 이상치 검사
check_columns = ['CO2ppm', 'Temp', 'PI_abs', 'Fv-Fm', 'Leaf_TPC', 'Root_TPC']

outlier_summary = {}
for name, df in datasets.items():
    outlier_summary[name] = {}
    print(f"\n[{name} 시나리오 이상치]")
    
    for col in check_columns:
        outliers, lower, upper = detect_outliers_iqr(df, col)
        outlier_count = len(outliers)
        outlier_percentage = (outlier_count / len(df)) * 100
        
        outlier_summary[name][col] = {
            'count': outlier_count,
            'percentage': outlier_percentage,
            'lower_bound': lower,
            'upper_bound': upper
        }
        
        if outlier_count > 0:
            print(f"  • {col}: {outlier_count}개 ({outlier_percentage:.1f}%)")
            print(f"    정상범위: [{lower:.2f}, {upper:.2f}]")
```

```python
# 이상치 시각화
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('주요 변수 이상치 분포', fontsize=16, fontweight='bold')

for idx, col in enumerate(check_columns):
    ax = axes[idx//3, idx%3]
    
    # 각 시나리오별 데이터와 이상치 표시
    for name, df in datasets.items():
        outliers, lower, upper = detect_outliers_iqr(df, col)
        
        # 전체 데이터 산점도
        ax.scatter(df.index, df[col], alpha=0.5, s=10, label=f'{name}')
        
        # 이상치 강조
        if len(outliers) > 0:
            ax.scatter(outliers.index, outliers[col], color='red', 
                      s=30, marker='x', alpha=0.8)
    
    ax.set_title(f'{col} 이상치')
    ax.set_xlabel('Index')
    ax.set_ylabel(col)
    ax.legend()
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# ====================================
# 6. SSP 시나리오별 비교 분석
# ====================================

print("="*60)
print("🔄 SSP 시나리오별 비교 분석")
print("="*60)

# 시나리오별 평균값 비교
comparison_df = pd.DataFrame()

for name, df in datasets.items():
    means = df.select_dtypes(include=[np.number]).mean()
    comparison_df[name] = means

# 주요 변수들만 선택
important_vars = ['CO2ppm', 'Temp', 'Humid', 'VPD', 'Chl_a', 'TChl', 
                 'PI_abs', 'Fv-Fm', 'Leaf_TPC', 'Root_TPC', 'Leaf_TFC', 'Root_TFC']

comparison_subset = comparison_df.loc[important_vars]

print("\n📊 시나리오별 평균값 비교")
print(comparison_subset.round(3))
```

```python
# 변화율 계산 (SSP1 대비)
print("\n📈 SSP1 대비 변화율 (%)")
change_rate = pd.DataFrame()

baseline = comparison_subset['SSP1']
for scenario in ['SSP3', 'SSP5']:
    change_rate[f'{scenario} vs SSP1'] = ((comparison_subset[scenario] - baseline) / baseline * 100)

print(change_rate.round(2))
```

```python
# 시나리오별 비교 시각화
fig, axes = plt.subplots(3, 4, figsize=(20, 15))
fig.suptitle('SSP 시나리오별 주요 변수 비교', fontsize=16, fontweight='bold')

for idx, var in enumerate(important_vars):
    ax = axes[idx//4, idx%4]
    
    # 막대 그래프로 비교
    values = [comparison_subset.loc[var, scenario] for scenario in ['SSP1', 'SSP3', 'SSP5']]
    colors = ['lightblue', 'lightcoral', 'lightgreen']
    bars = ax.bar(['SSP1', 'SSP3', 'SSP5'], values, color=colors, edgecolor='black')
    
    # 값 표시
    for bar, value in zip(bars, values):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height,
                f'{value:.2f}', ha='center', va='bottom', fontsize=9)
    
    ax.set_title(var)
    ax.set_ylabel('평균값')
    ax.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```
