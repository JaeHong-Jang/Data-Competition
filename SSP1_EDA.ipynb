```python
# 상관관계 행렬을 계산하고 절댓값을 취함
corr_matrix = df1.corr().abs()

# 상관관계 행렬을 시리즈(Series) 형태로 변환하여 정렬하기 쉽게 만듦
# unstack()을 통해 행렬을 계층적 인덱스를 가진 시리즈로 변환
sol = corr_matrix.unstack()

# kind="quicksort"를 사용하여 내림차순으로 정렬
sorted_sol = sol.sort_values(kind="quicksort", ascending=False)

# 자기 자신과의 상관관계(값이 1.0)를 제외하고,
# (A,B)와 (B,A) 중복 쌍을 제거하기 위해 2칸씩 건너뛰며 상위 6개 추출
top_pairs_raw = sorted_sol[sorted_sol < 1].iloc[::2].head(10)

print("📈 상관관계가 가장 강한 상위 6개 변수 쌍:")
print(top_pairs_raw)
print("-" * 50)


# --------------------------------------------------------------------------
# [3단계] 선택된 변수 쌍에 대한 산점도 생성 및 저장
# --------------------------------------------------------------------------
print("선택된 변수 쌍에 대한 산점도를 생성합니다...")

# items()를 사용하여 (변수1, 변수2) 쌍과 상관계수 값을 함께 가져와 반복
for (var1, var2), corr_value in top_pairs_raw.items():
    # 새로운 그래프 창 생성
    plt.figure(figsize=(10, 7))

    # seaborn의 regplot을 사용하여 산점도와 회귀선을 함께 그림
    sns.regplot(x=df1[var1], y=df1[var2],
                line_kws={"color": "red", "lw": 2}, # 회귀선 스타일 지정
                scatter_kws={'alpha': 0.6})         # 점들의 투명도 지정

    # 그래프 제목에 변수 이름과 상관 계수를 함께 표시
    plt.title(f"'{var1}'와 '{var2}'의 관계 (상관계수: {corr_value:.4f})", fontsize=16)
    plt.xlabel(var1, fontsize=12)
    plt.ylabel(var2, fontsize=12)
    plt.grid(True)
```

```python
# 상관관계 행렬을 계산하고, 모든 값을 양수로 만들기 위해 절댓값을 취함
corr_matrix = df1.corr().abs()

# 중복된 값과 자기 자신과의 상관관계(1.0)를 제거하기 위한 과정
# np.triu: 행렬에서 주대각선(k=1) 위쪽의 값만 남기고 나머지는 0으로 만드는 함수
# .where: 조건이 참인 위치의 값은 그대로 두고, 거짓인 위치는 다른 값(NaN)으로 채움
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# 행렬 형태의 데이터를 (변수1, 변수2, 상관계수) 형태의 목록으로 변환하고,
# 값이 없는(NaN) 항목은 제거
to_check = upper.unstack().dropna()

# 위에서 만든 목록에서 상관계수가 0.7보다 큰 항목만 필터링하고,
# 높은 순으로 정렬
high_corr_pairs = to_check[to_check > 0.7].sort_values(ascending=False)
```

```python
print("="*60)
print(" collinearity (상관계수 > 0.7)가 의심되는 변수 쌍 목록")
print(f"(총 {len(high_corr_pairs)}개)")
print("="*60)

if high_corr_pairs.empty:
    print("상관계수 0.7 이상인 변수 쌍이 없습니다.")
else:
    # 보기 좋게 출력
    print(high_corr_pairs)
```

```python
# 상관관계 행렬 계산
corr_matrix = df1.corr()

# 히트맵을 그릴 figure의 사이즈 설정
plt.figure(figsize=(20, 16))

# 상관계수의 절댓값이 0.7 미만인 값들은 그래프에 표시하지 않도록 '마스크(mask)' 생성
mask = np.abs(corr_matrix) < 0.7

# seaborn의 heatmap 함수를 사용하여 시각화
sns.heatmap(corr_matrix, 
            annot=False,        # 셀 안에 상관계수 값을 표시하지 않음 (변수가 많아 복잡해짐)
            cmap='viridis',     # 색상 팔레트 지정 (밝을수록 강한 양의 상관관계)
            mask=mask,          # 위에서 생성한 마스크 적용
            linewidths=.5)      # 각 셀을 구분하는 라인 추가

# 그래프 제목 및 축 레이블 설정
plt.title('상관계수 0.7 이상인 변수 쌍 히트맵', fontsize=20)
plt.xticks(rotation=45, ha='right') # x축 라벨이 겹치지 않도록 회전
plt.yticks(rotation=0)
plt.tight_layout() # 그래프 요소들이 겹치지 않도록 자동 조정
```

```python
# ====================================
# 2. 이상치 탐지 함수 정의
# ====================================

def detect_outliers_iqr(df, column):
    """IQR 방법으로 이상치 탐지"""
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# ====================================
# 3. df1의 전체 변수에 대한 이상치 탐지 실행
# ====================================

print("="*60)
print("🎯 df1(SSP1)의 전체 변수에 대한 이상치 탐지 (IQR 방법)")
print("="*60)

found_outlier = False

# df1의 모든 열(변수)에 대해 반복 (month는 제외)
for col in df1.columns:
    if col == 'month':
        continue # month 열은 건너뛰기

    # 이상치 탐지 함수 호출
    outliers, lower, upper = detect_outliers_iqr(df1, col)
    outlier_count = len(outliers)
    
    # 이상치가 발견된 경우에만 결과 출력
    if outlier_count > 0:
        outlier_percentage = (outlier_count / len(df1)) * 100
        print(f"  • {col}: {outlier_count}개 ({outlier_percentage:.1f}%)")
        print(f"    정상범위: [{lower:.2f}, {upper:.2f}]")
        found_outlier = True

if not found_outlier:
    print("  -> 발견된 이상치가 없습니다.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# 한글 폰트 설정
try:
    plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows
except:
    try:
        plt.rcParams['font.family'] = 'AppleGothic'  # Mac
    except:
        plt.rcParams['font.family'] = 'DejaVu Sans'  # Linux
        
plt.rcParams['axes.unicode_minus'] = False


def plot_monthly_avg_histogram(data, col):
    """월별 평균변화 히스토그램 형태로 그리기"""
    monthly_avg = data.groupby('month')[col].mean()
    
    plt.figure(figsize=(10, 6))
    plt.bar(monthly_avg.index, monthly_avg.values, color='skyblue', alpha=0.8)
    plt.title(f'{col} 월별 평균 변화 (히스토그램)', fontsize=14, fontweight='bold')
    plt.xlabel('월', fontsize=12)
    plt.ylabel(f'{col} 평균', fontsize=12)
    plt.grid(axis='y', alpha=0.3)
    plt.xticks(monthly_avg.index)
    plt.tight_layout()
    plt.show()

# 모든 컬럼 월별 평균변화 히스토그램 그리기
print("📊 모든 컬럼 월별 평균변화 히스토그램 생성!")
print(f"총 {len(df1.columns)}개 컬럼 분석")

for i, column in enumerate(df1.columns):
    print(f"\n[{i+1}/{len(df1.columns)}] {column} 월별 평균변화 생성 중...")
    plot_monthly_avg_histogram(df1, column)

print("✅ 모든 컬럼 월별 평균변화 히스토그램 완료!")
```
